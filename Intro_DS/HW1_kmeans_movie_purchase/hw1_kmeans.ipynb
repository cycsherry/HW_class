```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
c_data = pd.read_csv("cdata.csv")
```

```python
c_data=c_data[["x","y"]]
k=4
centroids=(c_data.sample(n=k))
#print(centroids)

diff=1
t=0
while diff!=0:
    c_data1=c_data
    i=1
    for index1,row_a in centroids.iterrows():
        D=[]
        for index2,row_b in c_data1.iterrows():
            d1=(row_a["x"]-row_b["x"])**2
            d2=(row_a["y"]-row_b["y"])**2
            d=np.sqrt(d1+d2)
            D.append(d)
        c_data[i]=D
        i=i+1
    #print(c_data)
    clu=[]
    for index,row in c_data.iterrows():
        mid=row[1]
        p=1
        for m in range(k):
            if row[m+1]<mid:
                mid=row[m+1]
                p=m+1
        clu.append(p)
    #print(clu)
    c_data["cluster1"]=clu
    cen_new=c_data.groupby("cluster1").mean()[["x","y"]]
    #print(cen_new)
    if t==0:
        diff=1
        t=t+1
    else:
        diff = (cen_new['x'] - centroids['x']).sum() + (cen_new['y'] - centroids['y']).sum()
        #print(diff.sum())
    centroids=c_data.groupby(["cluster1"]).mean()[["x","y"]]
    color=['blue','green','red',"yellow"]
    alp=["A","B","C","D"]
    for r in range(k):
        data=c_data[c_data["cluster1"]==r+1]
        plt.scatter(data["x"],data["y"],c=color[r])
    plt.scatter(centroids["x"],centroids["y"],c='black')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(labels=["A","B","C","D","centroids"],loc="upper right")
    plt.show()
```

```python
c_data = pd.read_csv("cdata.csv")
c_data=c_data[["x","y"]]
k1=list(range(2,51))
#print(c_data)
SSE=[]
for k in k1:
    centroids=(c_data.sample(n=k))
    #print(centroids)

    diff=1
    t=0
    while diff!=0:
        c_data1=c_data
        i=1
        for index1,row_a in centroids.iterrows():
            D=[]
            for index2,row_b in c_data1.iterrows():
                d1=(row_a["x"]-row_b["x"])**2
                d2=(row_a["y"]-row_b["y"])**2
                d=np.sqrt(d1+d2)
                D.append(d)
            c_data[i]=D
            i=i+1
        #print(c_data)
        clu=[]
        for index,row in c_data.iterrows():
            mid=row[1]
            p=1
            for m in range(k):
                if row[m+1]<mid:
                    mid=row[m+1]
                    p=m+1
            clu.append(p)
        #print(clu)
        c_data["cluster1"]=clu
        cen_new=c_data.groupby("cluster1").mean()[["x","y"]]
        #print(cen_new)
        if t==0:
            diff=1
            t=t+1
        else:
            diff = (cen_new['x'] - centroids['x']).sum() + (cen_new['y'] - centroids['y']).sum()
            #print(diff.sum())
        centroids=c_data.groupby(["cluster1"]).mean()[["x","y"]]
    #print(centroids)
    SSE1=[]
    for l in range(1,k+1):
        group=c_data.groupby("cluster1").get_group(l)
        cen=centroids.loc[l]
        for index,row in group.iterrows():
            d1=(cen["x"]-row["x"])**2
            d2=(cen["y"]-row["y"])**2
            SSE1.append(d1+d2)
   # print(sum(SSE1))
    SSE.append(sum(SSE1))
print(SSE)
plt.plot(k1,SSE,c='black')
plt.xlabel('number of k ')
plt.ylabel('Sum of square error')
plt.show()
```
