```python
def pct_by_group(data, feature, target):
    df = data.groupby([feature, target])[target].agg(['count'])
    temp = data.groupby([feature])[target].agg(['count'])
    df['%'] = 100*df.div(temp, level = feature).reset_index()['count'].values
    return df.reset_index()
```

```python
gender_count=pct_by_group(train,"Gender","Exited")
gender_count
```

資料處理
================

```python
#onehot binary
list_geoandgen = ['Geography', 'Gender']

train_onehot = pd.get_dummies(train, columns = list_geoandgen, prefix = list_geoandgen)
display(train_onehot.head())

test_onehot = pd.get_dummies(test, columns = list_geoandgen, prefix = list_geoandgen)
display(test_onehot.head())
```

```python
train_onehot=train_onehot.drop(columns=['RowNumber','CustomerId','Surname'])
test_onehot=test_onehot.drop(columns=['RowNumber','CustomerId','Surname'])
```

```python
features=list(train_onehot.drop(columns=["Exited"]))
target="Exited"
```

```python
sc=StandardScaler()

#標準化
train_onehot[features] = sc.fit_transform(train_onehot[features])
display(train_onehot)

#用train的數據對test標準化
test_onehot[features] = sc.transform(test_onehot[features])
display(test_onehot)
```

Logistic Regression
====================

```python
parameters = {'C': [0.01, 0.1, 1, 10],
              'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
              'max_iter': [50, 100, 150]}
LR= LogisticRegression(penalty = 'l2')
model_LR = GridSearchCV(LR, parameters, cv = 5, n_jobs = 10, verbose = 1).fit(train_onehot[features], train_onehot[target])
model_LR.best_params_
```

```python
LR_grid=LogisticRegression(C=10,max_iter=50,solver="newton-cg",random_state=42)
LR_grid.fit(train_onehot[features],train_onehot[target])
prediction_LR2=LR_grid.predict(test_onehot[features])
prediction_LR2
```

```python
upload_LRgrid=upload
upload_LRgrid["Exited"]=prediction_LR2
upload_LRgrid.to_csv("upload_gridLR.csv")
```

MLP
==============

```python
MLP=MLPClassifier(alpha=10)
mlp=MLP.fit(train_onehot[features],train_onehot[target])
prediction_MLP=MLP.predict(test_onehot[features])
prediction_MLP
```

```python
upload_MLP=upload
upload_MLP["Exited"]=prediction_MLP
upload_MLP.to_csv("upload_MLP.csv")
```
